```python
#Importando as Libs Necessárias
import pandas as pd
import numpy as np
```

```python
#bibliotecas para vizualização de dados
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importando o df
df_dados = pd.read_csv('train_u6lujuX_CVtuZ9i (1).csv')
df_dados.head()
```

```python
df_dados.info()
```

```python
#excluindo a coluna 'Loan_ID', pois não será usada. Separando 
#também as colunas categóricas e numéricas

df_dados1 = df_dados.drop(columns=['Loan_ID'])

colunas_categoricas = ['Gender', 'Married', 'Education', 'Dependents',
 'Self_Employed', 'Property_Area', 'Credit_History', 'Loan_Amount_Term']

colunas_numericas = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
```

```python
print(colunas_categoricas)
print(colunas_numericas)
```

```python
# Criando uma grade de subplots com 4 linhas e 2 colunas,
#resultando em um total de 8 subplots
#analisamos os valores das variáveis em relação a variável alvo
fig, axes = plt.subplots(4, 2, figsize = (12,15))

for i, col_cat in enumerate(colunas_categoricas):
    row , col = i // 2, i % 2
    sns.countplot(x = col_cat, data = df_dados, hue= 'Loan_Status', ax=axes[row, col])
```

```python
fig,axes = plt.subplots(1, 3, figsize= (17,5))
for i , col_num in enumerate(colunas_numericas):
    sns.boxplot(y = col_num, data = df_dados, x='Loan_Status', ax=axes[i])
```
